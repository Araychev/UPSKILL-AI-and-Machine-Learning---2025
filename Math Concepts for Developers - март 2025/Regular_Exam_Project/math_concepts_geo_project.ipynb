## 3. Theoretical Background

### 3.1 Geodesy

### What is Geodesy?

Geodesy is the science of accurately measuring and understanding the Earth's geometric shape, its orientation in space, and its gravity field. It provides the foundation for positioning, navigation, and mapping.

The word "geodesy" comes from the Greek words "geo" (Earth) and "daiein" (to divide), meaning "Earth dividing" or "Earth measuring".

Geodesy addresses three main components:

1. **Geometric Geodesy** â€“ studies the shape and size of the Earth. It provides coordinate systems (e.g., latitude, longitude, ellipsoidal height) and models the Earth as an ellipsoid or geoid.
2. **Gravimetric Geodesy** â€“ focuses on the Earth's gravity field and its variations. It is used in determining the geoid, which is a surface of equal gravitational potential and better represents "sea level" than an ellipsoid.
3. **Satellite Geodesy** â€“ uses satellites (like GPS, GLONASS, Galileo) for positioning and observing Earth deformations, sea level rise, and tectonic plate movements.

### Key Purposes of Geodesy

- Defining global and local coordinate systems (e.g., WGS84, UTM)
- Determining precise positions on Earth
- Measuring crustal movements (e.g., earthquakes, continental drift)
- Supporting mapping, navigation, and infrastructure
- Connecting surveying and geographic information systems (GIS)

### Real-World Applications

- **Surveying and construction** â€“ determining coordinates for property boundaries and engineering projects
- **Navigation and GPS** â€“ accurate positioning and routing
- **Earth observation and climate studies** â€“ measuring sea level rise, ice melt, land subsidence
- **Astronomy and space missions** â€“ linking Earth-based systems to celestial coordinates

Geodesy is essential for cartography and photogrammetry, as it defines the "true" shape of the Earth and how to relate images and maps to it.


### 3.2 Cartography

Cartography is the art and science of map-making. It involves representing the Earth's surfaceâ€”an inherently three-dimensional, curved objectâ€”onto two-dimensional maps. Because of the Earth's shape and complexity, creating accurate maps requires mathematical modeling and geospatial understanding.

Cartography links geodesy with visual representation, using coordinate systems and projections to translate real-world data into readable maps.

#### Main Components of Cartography

1. **Map Projections**
    - Mathematical methods for transforming the 3D surface of the Earth onto a 2D plane.
    - All projections involve some distortion of distance, direction, area, or shape.
    - Examples:
        - *Mercator* â€“ preserves direction (conformal), distorts area.
        - *Lambert Conformal Conic* â€“ used in aeronautical charts.
        - *Equal-area projections* â€“ preserve area, distort shape/distance.

2. **Coordinate Systems**
    - Maps use coordinate systems to identify locations.
    - Types:
        - *Geographic Coordinates* (latitude/longitude)
        - *Projected Coordinates* (e.g., UTM)
    - Each map must specify a **datum** and **projection** for coordinates to be meaningful.

3. **Scale**
    - Indicates the ratio between distances on the map and real-world distances.
    - Large-scale maps show smaller areas in more detail (e.g., 1:1,000).
    - Small-scale maps cover large areas with less detail (e.g., 1:1,000,000).

4. **Symbology and Design**
    - Visual representation of features using colors, symbols, and line styles.
    - Cartographic generalization simplifies complex real-world data to improve readability.

#### Role of Cartography in Geodesy

- Maps are the final products of geodetic surveys and photogrammetric missions.
- They help visualize measurements, errors, and transformations.
- Cartographic decisions affect how data is interpreted, especially when choosing projections or scales.

#### Mathematical Concepts in Cartography

- Projection formulas use trigonometry and calculus to transform geodetic coordinates.
- Distance and area distortion can be computed using differential geometry.
- Projections often require iterative numerical solutions (e.g., inverse projections).

Cartography is essential in transforming raw measurements into a human-readable, spatially meaningful format. Without it, geodetic data remains abstract and disconnected from practical applications like planning, navigation, or spatial analysis.


### 3.3 Photogrammetry

Photogrammetry is the science of obtaining reliable measurements and 3D information from photographs, especially aerial images. It is a bridge between geodesy, remote sensing, and computer vision.

The key idea is to use geometry to infer distances, positions, and shapes from two-dimensional images.

#### Types of Photogrammetry

1. **Aerial Photogrammetry**
    - Photographs taken from airplanes or drones.
    - Used for mapping, topographic surveys, and volume calculations.
  
2. **Terrestrial (Close-Range) Photogrammetry**
    - Images taken from ground level.
    - Used in architecture, forensics, archaeology, and engineering.

3. **Satellite Photogrammetry**
    - Based on satellite imagery, similar in principles to aerial photogrammetry.

#### Core Principles

- Based on **projective geometry**: the 3D world is projected onto a 2D image plane through a perspective transformation.
- Uses the **pinhole camera model** for mathematical modeling.
- Multiple overlapping images are used to extract depth and structure (stereoscopy).

#### Applications

- Creating orthophotos and 3D terrain models (DTMs/DSMs)
- Measuring distances, heights, and areas from images
- Generating input data for GIS systems
- Construction, mining, agriculture, and urban planning

#### Mathematical Concepts in Photogrammetry

      $$ y = yâ‚€ - f\frac{(r21(X - Xs) + r22(Y - Ys) + r23(Z - Zs))}{(r31(X - Xs) + r32(Y - Ys) + r33(Z - Zs))} $$
      

2. **Bundle Adjustment**
    - Simultaneous estimation of 3D point coordinates and camera positions/orientations using least squares optimization.

3. **Interior and Exterior Orientation**

### 4.2 Map Projection Equations

Projections provide mathematical methods to transform geographic coordinates onto a plane (map).

**Example 1: Mercator Projection (conformal, cylindrical)**

$$
\begin{aligned}
x &= R \cdot \lambda \\\\
y &= R \cdot \ln\left[\tan\left(\frac{\pi}{4} + \frac{\varphi}{2}\right)\right]
\end{aligned}
$$

This projection preserves angles but distorts areas towards the poles.

**Example 2: Lambert Conformal Conic Projection**

$$
\begin{aligned}
x &= \rho \cdot \sin[n(\lambda - \lambda_0)] \\\\
y &= \rho_0 - \rho \cdot \cos[n(\lambda - \lambda_0)]
\end{aligned}
$$

where $\rho, n, \rho_0$ are parameters derived from the standard parallels.

### 4.3 Triangulation / Bundle Adjustment (Basic Concepts)

In photogrammetry and classical geodetic measurements, **triangulation** is used to determine coordinates based on angle measurements or images of control points.

$$
\begin{aligned}

\end{aligned}
$$

These equations relate:
- $(X, Y, Z)$: object space point
- $(X_s, Y_s, Z_s)$: camera position
- $R$: rotation matrix describing orientation
- $(x, y)$: image coordinates

**Bundle adjustment**:  
Simultaneous refinement of all 3D point coordinates and camera parameters using least squares to minimize the difference between observed and calculated image points.

### 4.4 Error Propagation and Optimization

In geodesy, it is crucial to estimate how measurement errors affect results.

**Law of error propagation:**  
If a function $f$ depends on $n$ independent variables $x_i$ (with standard deviations $\sigma_{x_i}$), the variance of $f$ is:

$$
\sigma_f^2 = \sum_{i=1}^n \left( \frac{\partial f}{\partial x_i} \right)^2 \sigma_{x_i}^2
$$

**Least Squares Adjustment:**  
For linear observations:

$$
\mathbf{v} = A\mathbf{x} - \mathbf{l} \\\\
A^T A \mathbf{x} = A^T \mathbf{l} \\\\
\mathbf{x} = (A^T A)^{-1} A^T \mathbf{l}
$$

where:  
- $A$ is the design matrix  
- $\mathbf{x}$ is the vector of unknown parameters  
- $\mathbf{l}$ is the vector of observations  
- $\mathbf{v}$ is the vector of residuals

### 5.1 Geodetic to ECEF (Cartesian) Coordinates

#### Mathematical Background

Conversion from geodetic coordinates (latitude $\varphi$, longitude $\lambda$, ellipsoidal height $h$) to ECEF Cartesian coordinates $(X, Y, Z)$ uses the following equations:

$$
N = \frac{a}{\sqrt{1 - e^2 \sin^2 \varphi}}
$$

$$
\begin{aligned}
X &= (N + h) \cos\varphi \cos\lambda \\
Y &= (N + h) \cos\varphi \sin\lambda \\
Z &= [(1 - e^2)N + h] \sin\varphi
\end{aligned}
$$

where:
- $a$ is the semi-major axis (WGS84: $a = 6378137.0\,\mathrm{m}$)
- $b$ is the semi-minor axis (WGS84: $b = 6356752.314245\,\mathrm{m}$)
- $e$ is the eccentricity: $e = \sqrt{1 - (b^2/a^2)}$

```python

def geodetic_to_ecef(lat_deg, lon_deg, h, a=6378137.0, b=6356752.314245):
    """
    Converts geodetic coordinates (latitude, longitude in degrees, ellipsoidal height)
    to ECEF XYZ coordinates.

    Parameters:
        lat_deg (float or np.array): Latitude in degrees
        lon_deg (float or np.array): Longitude in degrees
        h (float or np.array): Height above ellipsoid in meters
        a (float): Semi-major axis (WGS84 by default)
        b (float): Semi-minor axis (WGS84 by default)

    Returns:
        X, Y, Z (float or np.array): ECEF coordinates in meters
    """
    lat = np.deg2rad(lat_deg)
    lon = np.deg2rad(lon_deg)
    e2 = 1 - (b**2)/(a**2)
    N = a / np.sqrt(1 - e2 * np.sin(lat)**2)
    X = (N + h) * np.cos(lat) * np.cos(lon)
    Y = (N + h) * np.cos(lat) * np.sin(lon)
    Z = ((1 - e2) * N + h) * np.sin(lat)
    return X, Y, Z
```

### 5.2 ECEF (Cartesian) to Geodetic Coordinates

#### Mathematical Background

Given ECEF coordinates $(X, Y, Z)$, we want to find geodetic latitude $\varphi$, longitude $\lambda$, and ellipsoidal height $h$.

Let:
- $a$ â€” semi-major axis (WGS84: $6378137.0\,\mathrm{m}$)
- $b$ â€” semi-minor axis (WGS84: $6356752.314245\,\mathrm{m}$)
- $e^2 = 1 - (b^2/a^2)$ â€” eccentricity squared

Longitude:
$$
\lambda = \arctan2(Y, X)
$$

Initial calculations:
$$
p = \sqrt{X^2 + Y^2}
$$

Iterative solution for latitude ($\varphi$):
1. Initial guess:
   $$
   \varphi_0 = \arctan\left(\frac{Z}{p \cdot (1 - e^2)}\right)
   $$
2. Iterate:
   $$
   N = \frac{a}{\sqrt{1 - e^2 \sin^2 \varphi}}
   $$
   $$
   h = \frac{p}{\cos\varphi} - N
   $$
   $$
   \varphi_{new} = \arctan\left(\frac{Z}{p \cdot (1 - \frac{e^2 N}{N + h})} \right)
   $$

Until $|\varphi_{new} - \varphi|$ is small (e.g. $10^{-12}$ radians).

Height:
$$
h = \frac{p}{\cos\varphi} - N
$$

```python
def ecef_to_geodetic(X, Y, Z, a=6378137.0, b=6356752.314245, tol=1e-12):
    """
    Converts ECEF (X,Y,Z) to geodetic coordinates (latitude, longitude in degrees, ellipsoidal height in meters).
    Uses iterative Bowring's method.
    """
    e2 = 1 - (b**2 / a**2)
    ep2 = (a**2 - b**2) / b**2
    p = np.sqrt(X**2 + Y**2)
    # Longitude
    lon = np.arctan2(Y, X)
    
    # Initial latitude approximation
    lat = np.arctan2(Z, p * (1 - e2))
    
    # Iterative improvement
    lat_prev = 0
    while np.abs(lat - lat_prev) > tol:
        lat_prev = lat
        N = a / np.sqrt(1 - e2 * np.sin(lat)**2)
        h = p / np.cos(lat) - N
        lat = np.arctan2(Z, p * (1 - e2 * N / (N + h)))

    # Final height calculation
    N = a / np.sqrt(1 - e2 * np.sin(lat)**2)
    h = p / np.cos(lat) - N
    
    lat_deg = np.rad2deg(lat)
    lon_deg = np.rad2deg(lon)
    return lat_deg, lon_deg, h
```

### 5.3 Surface Distance Between Two Points (Haversine Formula)

#### Mathematical Background

The Haversine formula is a well-known approach to calculate the great-circle distance \( d \) between two points on the Earth's surface, given their geographic coordinates (latitude and longitude). It assumes a spherical Earth (which is sufficiently accurate for many practical tasks).

Let:
- $\varphi_1, \lambda_1 $: coordinates of point 1 (latitude, longitude in radians)
- $\varphi_2, \lambda_2 $: coordinates of point 2 (latitude, longitude in radians)

The formulas are:

$$
a = \sin^2\left(\frac{\Delta\varphi}{2}\right) +
    \cos \varphi_1 \cdot \cos \varphi_2 \cdot \sin^2\left(\frac{\Delta\lambda}{2}\right)
$$

$$
d = 2R \cdot \arctan2\left(\sqrt{a}, \sqrt{1-a}\right)
$$

Where:
- $ \Delta\varphi = \varphi_2 - \varphi_1 $
- $\Delta\lambda = \lambda_2 - \lambda_1 $
-  R  = mean radius of Earth (6,371,000 meters by convention)

---

```python
def haversine_distance(lat1, lon1, lat2, lon2, radius=6371000):
    """
    Calculates the great-circle distance (in meters) between two points
    given by latitude and longitude in degrees.
    """
    lat1_rad, lon1_rad = np.radians(lat1), np.radians(lon1)
    lat2_rad, lon2_rad = np.radians(lat2), np.radians(lon2)
    dlat = lat2_rad - lat1_rad
    dlon = lon2_rad - lon1_rad
    
    a = np.sin(dlat / 2)**2 + \
        np.cos(lat1_rad) * np.cos(lat2_rad) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    return radius * c
```

```python


# WGS84 ellipsoid; UTM zone is automatically determined by longitude
def latlon_to_utm(lat, lon):
    """Converts latitude and longitude in degrees to UTM coordinates (E,N,Zone,Letter)."""
    # zone calculation
    zone_number = int((lon + 180) / 6) + 1
    is_northern = lat >= 0
    proj_str = f"+proj=utm +zone={zone_number} +datum=WGS84 +units=m +no_defs"
    proj = Proj(proj_str)
    easting, northing = proj(lon, lat)
    return easting, northing, zone_number, "N" if is_northern else "S"

def utm_to_latlon(easting, northing, zone, northern=True):
    """Converts UTM coordinates back to latitude and longitude."""
    proj_str = f"+proj=utm +zone={zone} +datum=WGS84 +units=m +no_defs"
    proj = Proj(proj_str)
    lon, lat = proj(easting, northing, inverse=True)
    return lat, lon
```

### 5.5 Area and Perimeter of a Polygon on Earthâ€™s Surface

#### Mathematical Background

##### (1) Area on a Plane ("Shoelace formula")

For small polygons (parcel, city block), where curvature is negligible and coordinates are projected (e.g. UTM):

$$
A = \frac{1}{2} \left| \sum_{i=1}^{n} (x_i y_{i+1} - x_{i+1} y_i) \right|
$$

Where:
- $(x_i, y_i)$: coordinates of the i-th vertex (UTM or local system)
- $n$: number of vertices (close the polygon: $x_{n+1} = x_1$, $y_{n+1} = y_1$)

##### (2) Perimeter

Sum of distances between consecutive vertices:

$$
P = \sum_{i=1}^{n} \text{distance}(i, i+1)
$$

##### (3) Area on a Sphere

If the vertices are given with geographic coordinates (lat, lon):

Spherical excess method (for simple polygons):

$$
A = |E| \cdot R^2
$$

Where $E$ = spherical excess (sum of interior angles minus $(n-2)\pi$), $R$ = radius of Earth.

*For practical coding we recommend using libraries for area on ellipsoid, or you can use a good spherical approximation for moderate areas.*

---

**(1) Shoelace formula (projected coordinates)**

```python
def polygon_area_utm(x, y):
    """
    Calculates area of a closed polygon using Shoelace formula.
    x, y: arrays (or lists) with projected coordinates (meters, must be closed: x[0]==x[-1]).
    Returns area in square meters (absolute value).
    """
    x = np.asarray(x)
    y = np.asarray(y)
    return 0.5 * np.abs(np.dot(x, np.roll(y, -1)) - np.dot(y, np.roll(x, -1)))
```

**(2) Perimeter (projected)**

```python
def polygon_perimeter_utm(x, y):
    """
    Calculates the perimeter of a polygon (in meters).
    x, y: arrays of projected coordinates.
    """
    x = np.asarray(x)
    y = np.asarray(y)
    return np.sum(np.sqrt(np.diff(x)**2 + np.diff(y)**2)) + np.sqrt((x[-1]-x[0])**2 + (y[-1]-y[0])**2)
```

**(3) Spherical area (geographic coordinates, small polygons)**

```python
def polygon_area_spherical(lats, lons, radius=6371000):
    """
    Approximate area of a polygon on sphere by summing spherical excesses (lats/lons in degrees).
    Returns area in square meters.
    Reference: https://trs.jpl.nasa.gov/bitstream/handle/2014/40409/09-4296.pdf (Equation 22)
    """
    lats_rad = np.radians(lats)
    lons_rad = np.radians(lons)
    n = len(lats_rad)
    # For each triangle (Assume convex polygon, simple case)
    area = 0
    for i in range(n):
        lat1, lon1 = lats_rad[i-1], lons_rad[i-1]
        lat2, lon2 = lats_rad[i], lons_rad[i]
        area += (lon2 - lon1) * (2 + np.sin(lat1) + np.sin(lat2))
    area = area * (radius**2) / 2
    return np.abs(area)
```

## 5.6 Photogrammetry: Basic Triangulation & Collinearity Equations

5.6.1 Mathematical Background
Collinearity equations express the geometric relationship between object space, camera position and orientation, and the image measurements:

$$
\begin{aligned}
x &= x_0 - f \frac{r_{11}(X-X_s) + r_{12}(Y-Y_s) + r_{13}(Z-Z_s)}{r_{31}(X-X_s) + r_{32}(Y-Y_s) + r_{33}(Z-Z_s)} \
y &= y_0 - f \frac{r_{21}(X-X_s) + r_{22}(Y-Y_s) + r_{23}(Z-Z_s)}{r_{31}(X-X_s) + r_{32}(Y-Y_s) + r_{33}(Z-Z_s)}
\end{aligned}
$$

ÐºÑŠÐ´ÐµÑ‚Ð¾:

$(X_s, Y_s, Z_s)$ â€“ ÐºÐ¾Ð¾Ñ€Ð´Ð¸Ð½Ð°Ñ‚Ð¸ Ð½Ð° ÐºÐ°Ð¼ÐµÑ€Ð°Ñ‚Ð°
$(X, Y, Z)$ â€“ Ð½ÐµÐ¸Ð·Ð²ÐµÑÑ‚Ð½Ð° Ñ‚Ð¾Ñ‡ÐºÐ°
$r_{ij}$ â€“ ÐµÐ»ÐµÐ¼ÐµÐ½Ñ‚ Ð¾Ñ‚ Ñ€Ð¾Ñ‚Ð°Ñ†Ð¸Ð¾Ð½Ð½Ð°Ñ‚Ð° Ð¼Ð°Ñ‚Ñ€Ð¸Ñ†Ð° (Ð¾Ñ‚ GPS/IMU)
$f$ â€“ Ñ„Ð¾ÐºÑƒÑÐ½Ð¾ Ñ€Ð°Ð·ÑÑ‚Ð¾ÑÐ½Ð¸Ðµ (mm)
$(x_0, y_0)$ â€“ Ð³Ð»Ð°Ð²Ð½Ð° Ñ‚Ð¾Ñ‡ÐºÐ° Ð½Ð° ÑÐµÐ½Ð·Ð¾Ñ€Ð° (ÑÑ€ÐµÐ´Ð° Ð½Ð° ÐºÐ°Ð´ÑŠÑ€Ð° Ð¸Ð»Ð¸ ÐºÐ°Ð»Ð¸Ð±Ñ€Ð¸Ñ€Ð°Ð½ Ð¾Ñ„ÑÐµÑ‚)


5.6.2 Python Implementation (Triangulation by Intersection of Rays)

```python
def triangulate_two_views(C1, R1, x1, C2, R2, x2, f):
    """
    C1, C2: Camera centers in world coordinates (3,)
    R1, R2: Camera rotation matrices (3,3) world-to-camera
    x1, x2: image coordinates relative to principal point (2,)
    f: focal length (same for both cameras, mm or px)
    Returns X: estimated 3D point (3,)
    """
    # Build normalized rays in camera coordinate systems
    X1_cam = np.array([x1[0], x1[1], -f])
    X1_cam = X1_cam / np.linalg.norm(X1_cam)
    X2_cam = np.array([x2[0], x2[1], -f])
    X2_cam = X2_cam / np.linalg.norm(X2_cam)
    # Rays in world coords
    ray1 = R1.T @ X1_cam
    ray2 = R2.T @ X2_cam
    # Find closest point of approach between two rays
    # X = C1 + s1*ray1, X = C2 + s2*ray2 â†’ minimize |(C1 + s1*ray1) - (C2 + s2*ray2)|
    A = np.stack([ray1, -ray2], axis=1)
    b = C2 - C1
    s, _, _, _ = np.linalg.lstsq(A, b, rcond=None)
    P1 = C1 + s[0]*ray1
    P2 = C2 + s[1]*ray2
    X = (P1 + P2)/2  # Midpoint
    return X
```

```python
# Camera 1: Center at (0,0,0), no rotation; point on the image: (5, 0)mm
f = 35  # mm standard lens
C1 = np.array([0, 0, 0])
R1 = np.eye(3)
x1 = np.array([5, 0])

# Camera 2: Center at (1,0,0), rotated 10 degrees in Y
C2 = np.array([1, 0, 0])
angle = np.deg2rad(10)
R2 = np.array([
    [np.cos(angle), 0, np.sin(angle)],
    [0, 1, 0],
    [-np.sin(angle), 0, np.cos(angle)]
])
x2 = np.array([-5, 0])

X_est = triangulate_two_views(C1, R1, x1, C2, R2, x2, f)
print("Estimated 3D point:", X_est)
```

5.7 Bundle Adjustment & Least Squares Optimization (Basic Example)

5.7.1 Theoretical Background

Bundle adjustment is an iterative method that refines camera parameters and 3D point locations by minimizing the reprojection error between observed and predicted image points.
In essence, it is a nonlinear least squares problem:

Given:

$n$ cameras (each with position, orientation, maybe internal parameters)
$m$ 3D points
Measured image coordinates for each point in each camera
We want to find the 3D coordinates and camera parameters that minimize the sum of squared differences between observed and modeled image points (â€œreprojection errorâ€):

$$
\text{min} \sum_{i,j} \left| \mathbf{x}{ij}^{\text{obs}} - \mathbf{x}{ij}^{\text{model}} (\theta_j, X_i) \right|^2
$$

where:

$\mathbf{x}_{ij}^{\text{obs}}$ â€“ measured image for point $i$ in camera $j$
$\mathbf{x}_{ij}^{\text{model}}$ â€“ model prediction based on current parameters
$\theta_j$ â€“ parameters of camera $j$ (external orientation, sometimes internals)
$X_i$ â€“ object coordinates of point $i$
In practice: The model is nonlinear (because of trigonometric/collinearity equations), solved with iterative methods (Levenberg-Marquardt, etc.). For most demo cases, a linear model or a single-step adjustment is shown!

5.7.2 Simple Least Squares Example

Example: Fitting a Line to 2D Points (the most basic case)
Suppose we have some measurements $(x_i, y_i)$. We want to find the line $y = ax + b$ that best fits the points in the least squares sense.
This is the classic overdetermined linear system:

$$
y = A\cdot p
$$

where

$A$ is a $N \times 2$ matrix: $A = [[x_1, 1], [x_2, 1], ..., [x_N, 1]]$
$p$ is the parameter vector $[a, b]^T$
The solution:

$$
p = (A^TA)^{-1}A^Ty
$$

**Python Implementation**

```python
def fit_line_least_squares(x, y):
    """
    Fits a line y = ax + b to data (x, y) using least squares.
    Returns: coefficients (a, b)
    """
    A = np.c_[x, np.ones_like(x)]
    p, res, rank, s = np.linalg.lstsq(A, y, rcond=None)
    return p  # [a, b]

# Example data
x_pts = np.array([0, 1, 2, 3])
y_pts = np.array([1.1, 1.9, 3.2, 4.1])

a, b = fit_line_least_squares(x_pts, y_pts)
print(f"Fitted line: y = {a:.3f}x + {b:.3f}")
```

**Visualization (Optional)**

```python

plt.scatter(x_pts, y_pts, label='Measurements')
plt.plot(x_pts, a*x_pts + b, 'r-', label='Least squares fit')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Least Squares Line Fitting Example")
plt.legend()
plt.grid(True)
plt.show()
```

5.7.3 Connection to Bundle Adjustment

In real bundle adjustment, the matrices ($A$, $y$) come from the linearization (Taylor expansion) of the full (nonlinear) photogrammetric model (collinearity equations), and the â€œbest fitâ€ is the minimal sum of all reprojection errors.

Modern photogrammetric and geodetic processing (including GNSS networks) always â€œclose the loopâ€ and estimate all unknowns by least squares or robustified variants.

For large problems, this is solved with iterative optimization algorithmsâ€”e.g., Levenberg-Marquardt (scipy.optimize.least_squares).

## 7. Engineering Geodesy

Engineering geodesy (applied/surveying geodesy) is a specialized field that provides geodetic support and high-precision spatial measurements for civil engineering, infrastructure, industrial construction, transport, energy, hydrotechnical, urban planning, mining, and monitoring projects. Its main objective is to ensure the accurate transfer and control of geodetic data and reference systems throughout all stages of the construction, maintenance, and monitoring of engineering structures.

---

### 7.1 Key Applications

- **Construction staking and layout:** Precise marking of axes, foundations, and structures according to design coordinates and plans.
- **Alignment for linear infrastructure:** Road, rail, pipeline, and tunnel alignment calculation, staking, and quality control (horizontal and vertical geometry, cross and longitudinal profiles).
- **Earthwork measurement:** Computing cut/fill (excavation and embankment) volumes from terrain models, profiles, or survey data.
- **Machine control and guidance:** Providing real-time positioning for construction machines and integrated GNSS/robotic systems.
- **Deformation monitoring:** Detecting, analyzing, and visualizing movement or settlement in buildings, bridges, dams, embankments, landslides, and mine pits.
- **Structural/industrial metrology:** High-precision measurements for industrial plants and prefabrication (steel/concrete alignment, settlements, crane rail geometry).
- **Engineering networks and coordinate systems:** Local project-specific coordinate transformations for construction sites, especially when integrating different survey data sources.

---

### 7.2 Typical Survey Tasks and Calculations

- **Stakeout calculations** (plotting design points or lines in the field from plans)
- **Transformation between local/site and national coordinate systems** (2D and 3D conformal, affine, Helmert transformations)
- **Calculation of arc/curve parameters** (roads, railways)
- **Cross-section/profile generation and volume calculation**
- **Digital terrain and surface modeling (DTM/TIN/Grid)**
- **Best-fit adjustment (least squares) for control networks and construction points**
- **Error analysis and propagation**
- **Monitoring data time series analysis and visualization**

---

### 7.3 Important Mathematical Methods

#### 7.3.1 Coordinate Transformations

- **2D Similarity (Helmert) Transformation:**  
  Used for transforming between local and global/project coordinate systems.  
  Equations:
  $$
  x' = a x - b y + t_x \\
  y' = b x + a y + t_y
  $$
  where $a = m \cos \theta$, $b = m \sin \theta$, $t_x, t_y$ = translation, $m$ = scale factor.

- **3D Similarity Transformation (7-parameter):**  
  Includes translation, rotation (3 axes), and uniform scale.

##### b) 2D/3D Helmert Transformation
- 2D Similarity (Helmert) Transformation
Calculates a transformation between two plane systems using the formula:
$$
\begin{bmatrix}
x' \
y'
\end{bmatrix}
= m
\begin{bmatrix}
\cos\theta & -\sin\theta \
\sin\theta & \cos\theta
\end{bmatrix}
\begin{bmatrix}
x \
y
\end{bmatrix}
+
\begin{bmatrix}
t_x \
t_y
\end{bmatrix}
$$

where:

m = scale
$\theta$ = rotation angle (in rad)
$t_x, t_y$ = translations along x and y
Used to attach measurement systems with arbitrary rotation, scale, and origin â€“ both in construction and in cadastre or photogrammetry.

- 3D Similarity (Helmert) Transformation
Generalizes to 7 parameters (three rotations, three translations, one scale):
$$
\begin{bmatrix}
X' \
Y' \
Z'
\end{bmatrix}
= m \cdot R \cdot
\begin{bmatrix}
X \
Y \
Z
\end{bmatrix}
+
\begin{bmatrix}
T_x \
T_y \
T_z
\end{bmatrix}
$$

R = 3D rotation matrix (Euler, Tait-Bryan or small angles).

Used in GNSS equations, registering local networks to a world system (ETRS89, WGS84), merging point clouds, etc.

 **Implementation of 2D Helmert (Similarity) Transformation in Python** 

```python
def similarity_2d(src, dst):
    """
    Computes 2D similarity "Helmert" transformation: scale, rotation, and translation.
    src, dst: Nx2 arrays of coordinates (src - local, dst - global).
    Returns:
        m (scale), theta (rotation [rad]), tx, ty (translations),
        residuals (difference vector)
    """
    src = np.asarray(src)
    dst = np.asarray(dst)
    assert src.shape == dst.shape

    # Centroids
    src_centroid = src.mean(axis=0)
    dst_centroid = dst.mean(axis=0)

    # Centralize
    src_c = src - src_centroid
    dst_c = dst - dst_centroid

    # Helmert components
    Sxy = np.sum(dst_c[:, 0] * src_c[:, 0] + dst_c[:, 1] * src_c[:, 1])
    Sxx = np.sum(src_c[:, 0]**2 + src_c[:, 1]**2)
    Syy = np.sum(dst_c[:, 1] * src_c[:, 0] - dst_c[:, 0] * src_c[:, 1])
    m = np.sqrt(Sxy**2 + Syy**2) / Sxx
    theta = np.arctan2(Syy, Sxy)

    # Transformation
    R = np.array([[np.cos(theta), -np.sin(theta)],
                  [np.sin(theta),  np.cos(theta)]])
    t = dst_centroid - m * (R @ src_centroid)

    # Transform src to dst_estimated
    src_tr = (m * (R @ src.T).T) + t
    residuals = dst - src_tr

    return m, theta, t[0], t[1], residuals
```

#### Example â€“ Practical Application
Imagine that you have:

local coordinates (e.g. vertices from a project axis/main grid)
and someone has measured them on the ground with BGS2005 (UTM!) coordinates

```python
# Local system (example - axis of a construction site)
local_points = np.array([
    [0,   0],
    [20,  0],
    [20, 25],
    [0,  25]
])

# Real (UTM) coordinates after translation, rotation and slight scaling
theta_true = np.radians(7.0)   # True rotation 7 degrees relative to the axis
scale_true = 1.012             # 1.2% difference in scale (e.g. from scale error or reduction)
tx_true = 695100               # real location (UTM easting)
ty_true = 4729600

R_true = np.array([
    [np.cos(theta_true), -np.sin(theta_true)],
    [np.sin(theta_true),  np.cos(theta_true)]
])
utm_points = (scale_true * (R_true @ local_points.T).T) + [tx_true, ty_true]
# (Add noise for even more realistic)

# Transformation:
m_est, theta_est, tx_est, ty_est, res = similarity_2d(local_points, utm_points)
print(f"Estimated scale: {m_est:.6f}\nEstimated rotation: {np.degrees(theta_est):.3f}Â°\nTranslation: {tx_est:.2f}, {ty_est:.2f}")
print(f"Residuals (in meters):\n{res}")
```

##### c) Best-Fit Line/Curve (Stakeout, Monitoring)


In engineering geodesy and surveying, the **best-fit line** (or in some cases, a best-fit curve) is essential for:
- Aligning measurement points for linear objects (roads, railways, tunnels, pipelines)
- Stakeout and as-built control (comparing designed and measured axes)
- Monitoring (detecting shifts or trends in time-series point data)

---

#### Mathematical Background

Given a sequence of measured points $ \{(x_i, y_i)\}_{i=1}^N $, the goal is to find the line $y = ax + b$ which minimizes the sum of squared residuals:
$$
\min_{a, b} \sum_{i=1}^N [y_i - (a x_i + b)]^2
$$
This is a classic least squares optimization, with analytic solution:
$$
a = \frac{N \sum x_i y_i - \sum x_i \sum y_i}{N \sum x_i^2 - (\sum x_i)^2}
$$
$$
b = \frac{\sum y_i - a \sum x_i}{N}
$$
In matrix form, for N points:
$$
\mathbf{y} = A \cdot \mathbf{p}, \quad A =
\begin{bmatrix}
x_1 & 1 \\
x_2 & 1 \\
\vdots & \vdots \\
x_N & 1
\end{bmatrix},
\quad
\mathbf{p} = \begin{bmatrix} a \\ b \end{bmatrix}
$$
The best-fit parameters:
$$
\mathbf{p}^* = (A^T A)^{-1} A^T \mathbf{y}
$$

For non-linear trends or vertical alignment (roads, rails), higher order polynomials $ y = a_2 x^2 + a_1 x + a_0 $, or parametric (e.g., circle, parabola), can be fitted.

---

#### Python Example: 2D Best-Fit Line

```python
# Simulate measured points along a line, with some random error (as in stakeout or monitoring)
np.random.seed(1)
x = np.linspace(0, 30, 15)
y_exact = 5 + 0.45 * x        # Design line
y_measured = y_exact + np.random.normal(scale=0.15, size=x.size)  # Measured with noise

# Linear least squares fit
A = np.c_[x, np.ones_like(x)]
p, res, rank, s = np.linalg.lstsq(A, y_measured, rcond=None)
a, b = p

print(f"Best-fit line: y = {a:.4f} x + {b:.4f}")
```

**Visualization**

```python
plt.figure(figsize=(9,5))
plt.scatter(x, y_measured, label='Measured points', color='tab:blue')
plt.plot(x, y_exact, 'g--', label='Design (projected) axis', lw=1.5)
plt.plot(x, a*x + b, 'r-',  label=f"Best-fit line\ny = {a:.3f} x + {b:.3f}")
for xi, yi, yf in zip(x, y_measured, a*x + b):
    plt.plot([xi, xi], [yi, yf], 'k:', lw=0.7)   # Residual lines
plt.xlabel("Distance (m) along axis")
plt.ylabel("Elevation or Y (m)")
plt.title("Best-Fit Line: Stakeout or Monitoring Example")
plt.legend(loc="best")
plt.grid()
plt.tight_layout()
plt.show()
```

#### Engineering Interpretation
- **For stakeout:** The best-fit (red) line gives you the true measured axis versus the project line. The residuals (vertical dashed) indicate as-built deviations.
- **For monitoring:** When fitting points from repeated measurements over time, the best-fit lineâ€™s slope can reveal trends (e.g., settlement, tilting).
---
#### For Higher-Order or Best-Fit Curve
Linear fitting can be extended:
$$
y = a_2 x^2 + a_1 x + a_0 \
A = \begin{bmatrix} x_1^2 & x_1 & 1 \ \cdots & \cdots & \cdots \ x_N^2 & x_N & 1 \end{bmatrix}
$$
Use:

```python
A_poly = np.c_[x**2, x, np.ones_like(x)]
coeffs, *_ = np.linalg.lstsq(A_poly, y_measured, rcond=None)
y_fit = coeffs[0]*x**2 + coeffs[1]*x + coeffs[2]
plt.plot(x, y_fit, color='r')
plt.scatter(x, y_measured, color='b')
plt.show()
```

### 7.5 Example: Stakeout Calculation (Polar - Cartesian)

```python
def stakeout_polar(base_x, base_y, direction, distance):
    """
    Computes coordinates of a stakeout point given base (x, y), horizontal angle (rad), and slope distance (m).
    """
    x = base_x + distance * np.cos(direction)
    y = base_y + distance * np.sin(direction)
    return x, y
```
